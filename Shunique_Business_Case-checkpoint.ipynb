```python
# Ensured "Date" is to_datetime for future indexing 
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
# Created datasets for each year 
df = pd.read_csv('transactions.csv', parse_dates=['Date'])
Q_2020 = df[(df['Date'] >= '2020/01/01') & (df['Date'] <= '2020-12-31')]
Q_2019 = df[(df['Date'] >= '2019/01/01') & (df['Date'] <= '2019-12-31')]
Q_2018 = df[(df['Date'] >= '2018/01/01') & (df['Date'] <= '2018-12-31')]
```

```python
# Created datasets for Q2's 
Q2_2020 = Q_2020[(Q_2020.Date >= '2020-04-01') & (Q_2020.Date <= '2020-06-30')]
Q2_2019 = Q_2019[(Q_2019.Date >= '2019-04-01') & (Q_2019.Date <= '2019-06-30')]
```

```python
Q2sales = pd.DataFrame(Q2_2020.groupby('name').sum()['cost'])
Q2sales.sort_values('cost', ascending=False)[:3]
```

# Answer 1a.
The best selling shoe for Q2 in 2020 was Mirak Womens Khaki Lace Up Waterproof Walking Boot 3. 

```python
Q2sales19 = pd.DataFrame(Q2_2019.groupby('name').sum()['cost'])
Q2sales19.sort_values('cost', ascending=False)[:3]
```

# Answer 1b.
